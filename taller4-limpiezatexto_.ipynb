# Punto 1: Pre-Procesamiento

- `[5 pts]` Leer el archivo `dialogos.csv` usando `pandas`
- `[10 pts]` Crear una nueva columna con el texto en minúscula, sin caracteres especiales ni números y sin palabras vacias
- `[7 pts]` Usando la nueva columna, ¿cuáles son las 10 palabras más usadas? 

## Solución

```python
dialogos = pd.read_csv("C:/Users/TRABAJO/Documents/Maestria Analistica Estrategica de Datos/II semestre/NLP/dialogos.csv")
dialogos
```

```python
stopwords_sp = stopwords.words('spanish')
```

```python
caracteres = "!¡.%,""'""" "'';:¿?_-"
numeros = "0123456789"
dialogos["minusculas"] = dialogos['Locución'].str.lower()
dialogos["caracteres"] = dialogos['minusculas'].apply(lambda x:''.join([caracter for caracter in x if caracter not in (caracteres)]))
dialogos["numeros"] = dialogos['caracteres'].apply(lambda x:''.join([numero for numero in x if numero not in (numeros)]))
dialogos["pre-procesado"] = dialogos['numeros'].apply(lambda x:' '.join([palabra for palabra in x.split() if palabra not in (stopwords_sp)]).lower())
dialogos.drop(['minusculas', 'caracteres', 'numeros'], axis=1, inplace = True)

#dialogos["vacias"].replace(r"[^a-zA-Z0-9]"," ", regex=True)#me elimina tambien acentos y ñ
```

```python
dialogos
```

```python
conteo = []
for i in dialogos['pre-procesado']:
    lista = i.split()
    
    for c in lista:
        lista = str(c)
        conteo.append(lista)
        #print(c, c.count(c))
#conteo2 = ' '.join(cadena for cadena in conteo)    
#conteo2
for palabra in conteo:
    c =Counter(conteo)
print(c.most_common(10))
```

# Punto 2: Representación vectorial

- `[15 pts]` Crear una bolsa de palabras (BoW) del corpus usando la columna pre-procesada (Usando la función de `sklearn`)
- `[10 pts]` ¿Cuántas palabras hay en el vocabulario? (Usando la función de `sklearn`)

## Solucion

```python
corpus = dialogos
count_vect = CountVectorizer(binary=True)
bow_rep = count_vect.fit_transform(dialogos['pre-procesado'].values)
print("Vocabulario: ", count_vect.vocabulary_)
print()

print(corpus.iloc[1]) #la posicion 0
print(bow_rep[1].toarray())
```

```python
dialogos['bow'] = [row for row in bow_rep.toarray()]
dialogos
```

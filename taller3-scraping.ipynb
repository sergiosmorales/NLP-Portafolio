## Encabezados

```python
link="https://es.wikipedia.org/wiki/"
animales=["Caballo", "Cebra", "Burro", "Cocodrilo", "Gallo", "Raton", "Conejo", "Tigre", "Puma", "Rana"]

for animal in animales:
    animals=link+str(animal)
    #print(animals)
    request = urllib.request.Request(animals, headers={'User-Agent' : 'Mozilla/5.0'})
    webpage = urllib.request.urlopen(request)
    source = webpage.read()
    webpage.close
    soup = bs.BeautifulSoup(source, 'html.parser')
    tittle = soup.find(id="firstHeading").text
    print(f"El link es: {animals} y su encabezado es: {tittle}")
```
